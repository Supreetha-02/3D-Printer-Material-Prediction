```python
Q1 = data['fare'].quantile(0.25)
Q3 = data['fare'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + whisker_width*IQR
data['fare']=np.where(data['fare']>upper_whisker,upper_whisker,np.where(data['fare']<lower_whisker,lower_whisker,data['fare']))
```

```python
sns.boxplot(data['fare'],data=data)
```

```python
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + whisker_width*IQR
data['age']=np.where(data['age']>upper_whisker,upper_whisker,np.where(data['age']<lower_whisker,lower_whisker,data['age']))
```

```python
sns.boxplot(data['age'],data=data)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

columns = ['sex', 'embarked','class', 'embark_town', 'deck', 'alone']

for col in columns:
    le.fit(data[col])
    data[col] = le.transform(data[col])
    
data.head()
```
